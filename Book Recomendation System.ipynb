```python
num_rating = ratings_with_books.groupby("title")["rating"].count().reset_index()
```

```python
final_rating = ratings_with_books.merge(num_rating,on="title")
```

```python

final_rating = final_rating[final_rating["num_of_rating"]>50]
```

```python
book_pivote=final_rating.pivot_table(values="rating",columns="user_id",index="title",fill_value=0)
```

```python
book_sparse = csr_matrix(book_pivote)
```

```python
from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(algorithm="brute")
```

```python
def recomend_book(book_name):
    book_id = np.where(book_pivote.index== book_name)[0][0]
    distance,suggestion = model.kneighbors(book_pivote.iloc[book_id].values.reshape(1,-1),n_neighbors=6)
    for i in range(len(suggestion)):
        books = book_pivote.index[suggestion[i]]
    for j in books:
        print(j)
```
